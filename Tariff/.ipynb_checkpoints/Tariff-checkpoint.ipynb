2.2 Количество сделанных звонков и израсходованных минут разговора по месяцам для каждого пользователя 

Известно, что в датафрейме присутствуют данные с января по декабрь 2018 года. Cгенерируем массив из 12ти месяцев. Добавим к нему столбец "key", заполненный единицами.

```python
month = pd.date_range('2018-01-01','2018-12-31',freq = "MS").to_frame()
month[0]= month[0].dt.month
month['key'] = 1 
month.columns = ['month','key']
month
```

Такой же ключевой столбец с единицами добавим в таблицу с пользователями. В ней же создадим столбец с месяцем регистрации пользователя. В таблице со звонками также создадим столбец с месяцем звонка. 

```python
users['key'] = 1
users['reg_month'] = users['reg_date'].dt.month
calls['month'] = calls['call_date'].dt.month
```

Помесячный отчет со звонками формируется объединением новой таблицы month c таблицей users по ключу key. 

```python
report = month.merge(users, on = 'key')
report.drop('key',axis = 1, inplace = True)
report.head()
```

Из отчета удаляем строки, в которых месяц меньше месяца регистрации пользователя. И дважды группируем датафрейм calls - звонки 1) по месяцу 2) по пользователю с подсчетом количества звонков. Объединяем сгруппированную таблицу calls с отчетом.

```python
report = report[report['month']>=report['reg_month']]
calls_month_count = calls.groupby(['month','user_id'])['id'].count().reset_index()
report_month_count = report.set_index(['month','user_id']).join(calls_month_count.set_index(['month','user_id']))
report_month_count = report_month_count['id'].to_frame().rename(columns = {'id':'calls'})
display(report_month_count)
```

Мы видим одно отсутствующее значение в столбце с количеством звонков в декабре. Появление таких "NaN" объясняется тем, что они к соответствующему месяцу перестали быть абонентами или не совершали в тот месяц звонков.

В таблице calls длительность звонков представлена в минутах с округлением до двух знаков после запятой, т.е c точностью до сотых долей минуты. Для определения количества минут, израсходованным пользователем по месяцам, округлим длительности звонков согласно условию проекта - до целых минут в бОльшую сторону. 

```python
calls['duration_rounded'] = calls['duration'].apply(np.ceil)
```

```python
calls_month_dur = calls.groupby(['month','user_id'])['duration_rounded'].sum().reset_index()
report_month_dur = report.set_index(['month','user_id']).join(calls_month_dur.set_index(['month','user_id']))
report_month_dur = report_month_dur['duration_rounded'].to_frame().rename(columns = {'duration_rounded':'duration, m'})
display(report_month_dur)
```

2.3 Количество сделанных звонков и израсходованных минут разговора по месяцам для каждого пользователя

```python
messages['month'] = messages['message_date'].dt.month
messages_month_count = messages.groupby(['month','user_id'])['id'].count().reset_index()
report_month_mes = report.set_index(['month','user_id']).join(messages_month_count.set_index(['month','user_id']))
report_month_mes = report_month_mes['id'].to_frame().rename(columns = {'id':'messages'})
display(report_month_mes)
```

2.4 Объем израсходованного траффика по месяцам 

Согласно условию проекта траффик округлим до целых гигабайт в бОльшую сторону

```python
internet['month'] = internet['session_date'].dt.month
internet_month = internet.groupby(['month','user_id'])['mb_used'].sum().reset_index()
internet_month['gb']=(internet_month['mb_used']/1024).apply(np.ceil)
report_month_internet = report.set_index(['month','user_id']).join(internet_month.set_index(['month','user_id']))
report_month_internet = report_month_internet['gb'].to_frame().rename(columns = {'gb':'internet, gb'})
display(report_month_internet)
```

2.5 Помесячную выручка с каждого пользователя 

Создадим общий отчет с длительностью звонков, количеством сообщений и объемом траффика с помощью объединения таблиц.

```python
final_report = report_month_mes.join(report_month_internet)
final_report = final_report.join(report_month_dur)
final_report = final_report.join(report_month_count)
final_report
```

```python
final_report = pd.DataFrame(final_report).reset_index()
final_report
```

Присоедим к отчету столбец с тарифом из таблицы "users".

```python
users_t = users[['user_id','tariff']]
users_t.set_index(['user_id'])
final_report = users_t.merge(final_report, on = "user_id")
final_report
```

```python
users_churn_month=users[['user_id','churn_date']]
users_churn_month['churn_month']=users_churn_month['churn_date'].dt.month
users_churn_month.drop(columns=['churn_date'])
users_churn_month.set_index(['user_id'])
final_report = users_churn_month.merge(final_report, on = "user_id")

final_report['churn_month']=final_report['churn_month'].fillna(0).astype(int)
final_report=final_report.loc[(final_report['month'] <= final_report['churn_month']) | (final_report['churn_month']==0)]
final_report
```

```python
values={'messages': 0, 'internet, gb' : 0, 'duration, m' : 0}
final_report=final_report.fillna(value=values)
final_report[final_report['messages'].isna() | final_report['internet, gb'].isna() | final_report['duration, m'].isna()]
```

Вычтем столбца "duration" бесплатные минуты, соответствующие тарифам. Если за месяц количество бесплатных минут не превышено, то в соответствующую ячейку записываем "0". 

```python
tariffs=tariffs.set_index('tariff_name')
sm_d=tariffs.loc['smart','minutes_included']
ult_d=tariffs.loc['ultra', 'minutes_included']
final_report.loc[(final_report['duration, m'] <= sm_d) & (final_report['tariff'] == 'smart'), 'extra_dur'] = 0
final_report.loc[(final_report['duration, m'] > sm_d) & (final_report['tariff'] == 'smart'), 'extra_dur'] = final_report['duration, m']-sm_d
final_report.loc[(final_report['duration, m'] <= ult_d) & (final_report['tariff'] == 'ultra'), 'extra_dur'] = 0
final_report.loc[(final_report['duration, m'] > ult_d) & (final_report['tariff'] == 'ultra'), 'extra_dur'] = final_report['duration, m']-ult_d
```

```python
final_report
```

Отсутствующие значeния в столбцах обусловлены тем, что в конкретный месяц пользователь не пользовался услугами. На данном этапе с ними ничего не делаем.

Вычтем бесплатный лимит из количества сообщений и интернет траффика. 

```python
sm_m=tariffs.loc['smart','messages_included']
ult_m=tariffs.loc['ultra', 'messages_included']
sm_gb=tariffs.loc['smart','mb_per_month_included']/1024
ult_gb=tariffs.loc['ultra', 'mb_per_month_included']/1024
final_report.loc[(final_report['messages'] <= sm_m) & (final_report['tariff'] == 'smart'), 'extra_mes'] = 0
final_report.loc[(final_report['messages'] > sm_m) & (final_report['tariff'] == 'smart'), 'extra_mes'] = final_report['messages']-sm_m
final_report.loc[(final_report['messages'] <= ult_m) & (final_report['tariff'] == 'ultra'), 'extra_mes'] = 0
final_report.loc[(final_report['messages'] > ult_m) & (final_report['tariff'] == 'ultra'), 'extra_mes'] = final_report['messages']-ult_m
final_report.loc[(final_report['internet, gb'] <= sm_gb) & (final_report['tariff'] == 'smart'), 'extra_inet'] = 0
final_report.loc[(final_report['internet, gb'] > sm_gb) & (final_report['tariff'] == 'smart'), 'extra_inet'] = final_report['internet, gb']-sm_gb
final_report.loc[(final_report['internet, gb'] <= ult_gb) & (final_report['tariff'] == 'ultra'), 'extra_inet'] = 0
final_report.loc[(final_report['internet, gb'] > ult_gb) & (final_report['tariff'] == 'ultra'), 'extra_inet'] = final_report['internet, gb']-ult_gb
```

```python
final_report
```

Определим выручку за услуги сверх тарифа - запишем ее в три новых столбца.

```python
smart_gb_price = tariffs.loc['smart','rub_per_gb'] 
ultra_gb_price = tariffs.loc['ultra','rub_per_gb']
smart_dur_price = tariffs.loc['smart','rub_per_minute'] 
ultra_dur_price = tariffs.loc['ultra','rub_per_minute']
smart_mes_price = tariffs.loc['smart','rub_per_message'] 
ultra_mes_price = tariffs.loc['ultra','rub_per_message']
final_report.loc[(final_report['tariff'] == 'smart'), 'dur_revenue'] = final_report['extra_dur']*smart_dur_price
final_report.loc[(final_report['tariff'] == 'ultra'), 'dur_revenue'] = final_report['extra_dur']*ultra_dur_price
final_report.loc[(final_report['tariff'] == 'smart'), 'mes_revenue'] = final_report['extra_mes']*smart_mes_price
final_report.loc[(final_report['tariff'] == 'ultra'), 'mes_revenue'] = final_report['extra_mes']*ultra_mes_price
final_report.loc[(final_report['tariff'] == 'smart'), 'inet_revenue'] = final_report['extra_inet']*smart_gb_price
final_report.loc[(final_report['tariff'] == 'ultra'), 'inet_revenue'] = final_report['extra_inet']*ultra_gb_price
```

```python
final_report
```

Заполним пропуски в столбцах c количеством сообщений, интернет-траффика и длительностью разговоров

Создадим новый столбец с суммарной выручкой от каждого пользователя по месяцам: просуммируем абонентскую плату с платами вне тарифных лимитов.

```python
final_report.loc[(final_report['tariff'] == 'smart'), 'revenue'] = final_report['dur_revenue']+final_report['mes_revenue']+final_report['inet_revenue']+550
final_report.loc[(final_report['tariff'] == 'ultra'), 'revenue'] = final_report['dur_revenue']+final_report['mes_revenue']+final_report['inet_revenue']+1950
```

```python
final_report[final_report['messages'].isna() | final_report['internet, gb'].isna() | final_report['duration, m'].isna()]
```

### 4. Проверка гипотез
4.1 Cредняя выручка пользователей тарифа "смарт" и "ультра" отличаются 

Нулевая гипотеза Ho: "Средняя выручка пользователей тарифов "смарт" и "ультра" равна". Отсутствующие значения в соответствующем датафрейме выкинем. Проведем t-тест о равенстве средних двух генеральных совокупностей.

```python
smart_monthly_revenue = smart_monthly['revenue'].to_frame()
ultra_monthly_revenue = ultra_monthly['revenue'].to_frame()
results = st.ttest_ind(smart_monthly_revenue, ultra_monthly_revenue)
print('p-значение:', results.pvalue)
```

Даже если принять критичекий уровень статистической значимости (альфа) равным 1%, он не превышается. Нулевую гипотезу отвергаем. Следовательно, гипотеза "средняя выручка пользователей тарифа "смарт" и "ультра" отличаются" не отвергается. 

4.2 Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

Нулевая гипотеза: "Cредняя за месяц выручка пользователей из Москвы равна выручке пользователей из других регионов". Присоедим к отчету столбец с городом из таблицы "users". 

```python
users_city = users[['user_id','city']]
users_city.set_index(['city'])
final_report = users_city.merge(final_report, on = "user_id")
```

Проведем t-тест о равенстве средних двух генеральных совокупностей.

```python
moscow_revenue = final_report.loc[final_report['city'] == "Москва"]['revenue'].to_frame()
other_cities_revenue = final_report.loc[final_report['city'] != "Москва"]['revenue'].to_frame()
results = st.ttest_ind(moscow_revenue, other_cities_revenue)
print('p-значение:', results.pvalue)
```

Даже если принять критичекий уровень статистической значимости (альфа) равным 5%, он будет превышен. Нулевая гипотеза не отвергается.

В задании присутствует вопрос "Какой тариф лучше?". При прочих равных условиях (стоимость привлечения абонента, номинальная себестоимость услуг) лучше тариф, для которого средняя (или медианная) месячная выручка с одного абонента - выше. 

```python
smart_revenue_mean = final_report.loc[final_report['tariff'] == "smart"]['revenue'].mean()
ultra_revenue_mean = final_report.loc[final_report['tariff'] == "ultra"]['revenue'].mean()
smart_revenue_median = final_report.loc[final_report['tariff'] == "smart"]['revenue'].median()
ultra_revenue_median = final_report.loc[final_report['tariff'] == "ultra"]['revenue'].median()
```

```python
print('Средняя месячная выручка от одного пользователя тарифа "Smart"',round(smart_revenue_mean,0) ,'р')
print('Средняя месячная выручка от одного пользователя тарифа "Ultra"',round(ultra_revenue_mean,0) ,'р')
print('Медианная месячная выручка от одного пользователя тарифа "Smart"',round(smart_revenue_median,0) ,'р')
print('Медианная месячная выручка от одного пользователя тарифа "Ultra"',round(ultra_revenue_median,0) ,'р')
```

Как по средней, так и по медианной выручке на одного пользователя за месяц тариф "Ultra" опережает тариф "Smart". Соответственно, если ворпрос формулировать как "какой тариф лучше", то лучше тариф "Ultra".
